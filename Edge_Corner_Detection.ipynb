**Canny Edge Detection**

```
!wget -O egy_saudi_match.jpg https://raw.githubusercontent.com/heshameraqi/Image_Processing_Computer_Vision_1/main/egy_saudi_match.jpg

import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('egy_saudi_match.jpg', 0)
edges = cv2.Canny(img,100,200)

# Visualize
fig, ax = plt.subplots(1,2, figsize=(10,6))
ax[0].title.set_text('Original Image')
ax[0].imshow( img, cmap='gray' )
ax[1].title.set_text('Edge Image')
ax[1].imshow( edges, cmap='gray' )
plt.tight_layout()
```

**Harris Corner Detection**

```
!wget -O Chessboard.jpeg https://raw.githubusercontent.com/heshameraqi/Image_Processing_Computer_Vision_1/main/Chessboard.jpeg

import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('Chessboard.jpeg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

# Find Harris corners to detect the corners with appropriate values as input parameters
dest = cv2.cornerHarris(gray, 2, 5, 0.07) 
  
# Dilated the resulted corners corners and show them on the original image
dest = cv2.dilate(dest, None)
img[dest > 0.01 * dest.max()] = [0, 0, 255] 

# Visualize
fig, ax = plt.subplots(1,2, figsize=(10,6))
ax[0].title.set_text('Original Image')
ax[0].imshow( gray, cmap='gray' )
ax[1].title.set_text('Edge Image')
ax[1].imshow( cv2.cvtColor(img, cv2.COLOR_BGR2RGB) )
plt.tight_layout()
```
